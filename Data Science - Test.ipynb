```python

def evaluate_data_completeness(file_path):
    df = pd.read_csv(file_path)
    total_rows = len(df)
    missing_values = df.isnull().sum().sum()
    rows_with_missing_values = df.isnull().any(axis=1).sum()
    completeness_percentage = ((total_rows * len(df.columns) - missing_values) / (total_rows * len(df.columns))) * 100
    rows_with_missing_percentage = (rows_with_missing_values / total_rows) * 100
    return {
        'file': file_path,
        'total_rows': total_rows,
        'missing_values': missing_values,
        'rows_with_missing_values': rows_with_missing_values,
        'completeness_percentage': completeness_percentage,
        'rows_with_missing_percentage': rows_with_missing_percentage
    }

def analyze_folder(folder_path):
    results = []
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(folder_path, file_name)
            result = evaluate_data_completeness(file_path)
            results.append(result)
    return results

def plot_completeness(results):
    for result in results:
        labels = ['Complete', 'Missing']
        sizes = [result['completeness_percentage'], 100 - result['completeness_percentage']]
        colors = ['skyblue', 'lightcoral']
        explode = (0.1, 0)  # explode the 1st slice

        plt.figure(figsize=(8, 8))
        plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
                shadow=False, startangle=140)
        plt.title(f"Data Completeness of {os.path.basename(result['file'])}\nTotal Rows: {result['total_rows']}, Missing Values: {result['missing_values']}, Rows with Missing Values: {result['rows_with_missing_values']} ({result['rows_with_missing_percentage']:.2f}%)")
        plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.show()

folder_path = 'data'
evaluation_results = analyze_folder(folder_path)
plot_completeness(evaluation_results)
```
